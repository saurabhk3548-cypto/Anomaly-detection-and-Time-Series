Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Answer:Anomaly Detection is a data analysis technique used to identify unusual patterns, observations, or data points that do not conform to expected behavior. These unusual data points are called anomalies or outliers. Detecting such anomalies is important because they often indicate critical issues like fraud, system failure, or data errors.

Anomaly detection is widely used in fields such as banking (fraud detection), healthcare (disease diagnosis), cybersecurity (intrusion detection), and manufacturing (fault detection).

Types of Anomalies

1. Point Anomalies

A point anomaly occurs when a single data point is significantly different from the rest of the dataset.

Example:

In a dataset of daily expenses where most values are between ₹500 and ₹2000, if one transaction is ₹50,000, it is considered a point anomaly.

Explanation:

Here, only one data point deviates from the normal pattern, making it easy to detect.

2. Contextual Anomalies

A contextual anomaly occurs when a data point is abnormal in a specific context but may be normal in another context. The context could be time, location, or any other condition.

Example:

A temperature of 10°C is normal in winter but abnormal in summer.

Explanation:

The anomaly depends on the situation or environment. Without context, the data point may not appear unusual.

3. Collective Anomalies

A collective anomaly occurs when a group of data points together behaves abnormally, even though individual data points may appear normal.

Example:

A series of small transactions happening continuously from a bank account within a short time may indicate fraudulent activity.

Explanation:

Each transaction may look normal individually, but the pattern formed by the group is unusual.

Conclusion

Anomaly detection is a crucial technique for identifying irregular patterns in data. Understanding the different types—point, contextual, and collective anomalies—helps in analyzing data more effectively and taking appropriate actions in real-world applications.


Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Answer:Comparison of Isolation Forest, DBSCAN, and Local Outlier Factor (LOF)

Anomaly detection can be performed using different algorithms depending on the data structure and problem type. Three commonly used methods are Isolation Forest, DBSCAN, and Local Outlier Factor (LOF). Each follows a different approach and is suitable for different use cases.

1. Isolation Forest

Approach:

Isolation Forest is an ensemble-based algorithm that isolates anomalies instead of profiling normal data. It works by randomly selecting features and splitting data points. Anomalies are easier to isolate and therefore require fewer splits.

Key Idea:

Anomalies are rare and different, so they get isolated quickly.

Suitable Use Cases:

High-dimensional datasets

Fraud detection

Network intrusion detection

When data has many irrelevant features

Advantages:

Fast and efficient

Works well with large datasets

No need for distance calculation

Limitations:

Less effective when anomalies are not clearly separable

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:

DBSCAN is a clustering algorithm that groups data points based on density. Points in low-density regions are treated as anomalies (noise), while dense regions form clusters.

Key Idea:

Normal data forms dense clusters; anomalies lie in sparse regions.

Suitable Use Cases:

Spatial data analysis

Clustering with arbitrary shapes

Detecting noise in datasets

Geographic or sensor data

Advantages:

Does not require specifying number of clusters

Can detect arbitrarily shaped clusters

Naturally identifies outliers

Limitations:

Sensitive to parameter selection (epsilon and minPts)

Struggles with varying densities

3. Local Outlier Factor (LOF)

Approach:

LOF measures the local density of a data point compared to its neighbors. If a point has significantly lower density than its neighbors, it is considered an anomaly.

Key Idea:

Anomalies are points with lower density than their surrounding neighbors.

Suitable Use Cases:

Detecting local anomalies

Datasets with varying density

Credit card fraud detection

Intrusion detection

Advantages:

Captures local structure of data

Works well when density varies

Limitations:

Computationally expensive

Sensitive to number of neighbors parameter

Comparison Table:

| Feature           | Isolation Forest       | DBSCAN                       | LOF                       |
| ----------------- | ---------------------- | ---------------------------- | ------------------------- |
| Approach          | Random partitioning    | Density-based clustering     | Local density comparison  |
| Detects Anomalies | Easy-to-isolate points | Low-density (noise) points   | Low-density vs neighbors  |
| Works with        | High-dimensional data  | Spatial / clustered data     | Local density-based data  |
| Parameters        | Few                    | Epsilon, MinPts              | Number of neighbors       |
| Speed             | Fast                   | Moderate                     | Slow (for large datasets) |
| Best Use Case     | Large datasets, fraud  | Clustering + noise detection | Local anomaly detection   |

conclusion:

Isolation Forest is best for large and high-dimensional datasets, DBSCAN is ideal for clustering and detecting noise in spatial data, and LOF is suitable for identifying local anomalies where density varies. The choice of method depends on the nature of the dataset and the problem being solved.

Question 3: What are the key components of a Time Series? Explain each with one example.

Answer: Key Components of a Time Series

A time series is a sequence of data points collected or recorded at regular time intervals. To analyze time series data effectively, it is important to understand its main components.

1. Trend (T)

Definition:

Trend represents the long-term movement or direction in the data over a period of time. It shows whether the data is increasing, decreasing, or remaining constant.

Example:

The gradual increase in a company’s sales revenue over several years indicates an upward trend.

2. Seasonality (S)

Definition:

Seasonality refers to repeating patterns or fluctuations that occur at regular intervals due to seasonal factors such as months, quarters, or days.

Example:

Ice cream sales increase every summer and decrease in winter, showing a seasonal pattern.

3. Cyclical Component (C)

Definition:

Cyclical variations are long-term fluctuations that occur due to economic or business cycles. These do not follow a fixed period like seasonality.

Example:

Economic growth followed by recession and recovery phases in a country’s GDP over several years.

4. Irregular Component (I) / Noise

Definition:

This component represents random, unpredictable variations in the data that cannot be explained by trend, seasonality, or cycles.

Example:

A sudden drop in sales due to an unexpected event like a natural disaster or pandemic.

Summary

| Component   | Description                              | Example                     |
| ----------- | ---------------------------------------- | --------------------------- |
| Trend       | Long-term direction of data              | Increasing sales over years |
| Seasonality | Regular repeating pattern                | High sales in summer        |
| Cyclical    | Long-term fluctuations (no fixed period) | Economic cycles             |
| Irregular   | Random, unpredictable variations         | Sudden sales drop           |

Conclusion

Time series data is made up of trend, seasonal, cyclical, and irregular components. Understanding these components helps in better analysis, forecasting, and decision-making.



Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Answer:Stationarity in Time Series

A time series is said to be stationary if its statistical properties remain constant over time. This means that the mean, variance, and autocorrelation of the series do not change with time.

In simple terms, a stationary time series does not show trends, seasonal patterns, or changing variability.

Characteristics of a Stationary Series

Constant mean over time

Constant variance

No trend or seasonality

Autocorrelation depends only on lag, not on time

Example:

A dataset where values fluctuate randomly around a constant average (e.g., white noise).

Testing for Stationarity

1. Visual Inspection
   
Plot the time series

Check for trends, seasonality, or changing variance

Example:

If the graph shows an upward trend, the series is non-stationary.

2. Statistical Tests
   
a) Augmented Dickey-Fuller (ADF) Test

Most commonly used test

Null hypothesis: Series is non-stationary

If p-value < 0.05 → Series is stationary

b) KPSS Test

Null hypothesis: Series is stationary

If p-value < 0.05 → Series is non-stationary

Transforming Non-Stationary Series into Stationary

1. Differencing
   
Subtract previous value from current value

Removes trend

Formula (First-order differencing):

Yt′=Yt−Yt−1

Example:

If sales are increasing every month, differencing removes this upward trend.

2. Log Transformation
3. 
Reduces variance and stabilizes data

Example:

Used when data shows exponential growth (e.g., stock prices).

3. Detrending
4. 
Remove trend using regression or moving averages

Example:

Subtracting the trend line from the data.

4. Seasonal Adjustment
   
Remove seasonal patterns

Example:

Subtract average monthly values to remove seasonality.

Summary

| Aspect                | Description                                                  |
| --------------------- | ------------------------------------------------------------ |
| Stationary Series     | Constant mean, variance, no trend                            |
| Non-Stationary Series | Shows trend, seasonality, or changing variance               |
| Tests                 | ADF, KPSS                                                    |
| Transformations       | Differencing, Log transform, Detrending, Seasonal adjustment |

Conclusion

Stationarity is a crucial concept in time series analysis because many models (like ARIMA) require stationary data. Non-stationary data can be transformed into stationary form using techniques like differencing and transformation, enabling accurate modeling and forecasting.

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Answer:Difference between AR, MA, ARIMA, SARIMA, and SARIMAX Models

Time series forecasting uses various models depending on the nature of the data. The most commonly used models are AR, MA, ARIMA, SARIMA, and SARIMAX, each with different structures and applications.

1. AR (AutoRegressive) Model

Structure:

The AR model predicts the current value based on its past values.

Represented as: AR(p), where p = number of lag observations

Uses past values (lags) of the time series

Example:

Today’s stock price depends on the prices of previous days.

Application:

When data shows strong correlation with past values

Short-term forecasting

2. MA (Moving Average) Model

Structure:

The MA model uses past forecast errors (residuals) to predict current values.

Represented as: MA(q), where q = number of lagged errors

Uses error terms instead of actual past values

Example:

Prediction depends on previous errors in forecasting.

Application:

When data has random shocks or noise

Smoothing short-term fluctuations

3. ARIMA (AutoRegressive Integrated Moving Average)

Structure:

ARIMA combines AR and MA models along with differencing to make the series stationary.

Represented as: ARIMA(p, d, q)

p: AR terms

d: differencing order

q: MA terms

Example

Used for non-stationary data like stock prices after applying differencing.

Application:

General-purpose time series forecasting

Works well for non-seasonal data

4. SARIMA (Seasonal ARIMA)

Structure:

SARIMA extends ARIMA by adding seasonal components.

Represented as: SARIMA(p, d, q)(P, D, Q, s)

Seasonal parameters (P, D, Q)

s: seasonal period

Example

Monthly sales data with yearly seasonality (s = 12).

Application:

Data with clear seasonal patterns

Retail sales, weather data

5. SARIMAX (Seasonal ARIMA with Exogenous Variables)

Structure:

SARIMAX is an extension of SARIMA that includes external (exogenous) variables.

Represented as: SARIMAX(p, d, q)(P, D, Q, s) + external variables

Example:

Sales prediction using past sales data along with advertising spend or promotions.

Application:

When external factors influence the time series

Economic forecasting, demand forecasting

Comparison Table

| Model   | Structure                   | Handles Seasonality | Uses External Variables | Use Case                   |
| ------- | --------------------------- | ------------------- | ----------------------- | -------------------------- |
| AR      | Uses past values            | No                  | No                      | Simple time series         |
| MA      | Uses past errors            | No                  | No                      | Noise smoothing            |
| ARIMA   | AR + MA + Differencing      | No                  | No                      | Non-stationary data        |
| SARIMA  | ARIMA + Seasonal terms      | Yes                 | No                      | Seasonal data              |
| SARIMAX | SARIMA + External variables | Yes                 | Yes                     | Data with external factors |

Conclusion:

AR and MA are basic models, while ARIMA combines both for non-stationary data. SARIMA handles seasonal patterns, and SARIMAX further improves forecasting by including external variables. The choice of model depends on whether the data has trend, seasonality, or external influences.

```python
#Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
#(Include your Python code and output in the code box below.)

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Create synthetic taxi-like dataset
np.random.seed(42)
data = {
    'fare_amount': np.random.normal(15, 5, 500),
    'trip_distance': np.random.normal(5, 2, 500)
}

df = pd.DataFrame(data)

# Add some anomalies manually
df.loc[::50, 'fare_amount'] = df['fare_amount'] * 5
df.loc[::50, 'trip_distance'] = df['trip_distance'] * 0.2

# Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Isolation Forest
model = IsolationForest(contamination=0.05, random_state=42)
df['anomaly'] = model.fit_predict(scaled_data)
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})

print("Total anomalies detected:", df['anomaly'].sum())

# Plot
plt.figure(figsize=(8,6))

plt.scatter(df[df['anomaly'] == 0]['trip_distance'],
            df[df['anomaly'] == 0]['fare_amount'],
            label='Normal', alpha=0.5)

plt.scatter(df[df['anomaly'] == 1]['trip_distance'],
            df[df['anomaly'] == 1]['fare_amount'],
            color='red', label='Anomaly')

plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.title('Isolation Forest - Anomaly Detection')
plt.legend()
plt.show()
```

```python
#Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.
#(Include your Python code and output in the code box below.)

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

import warnings
warnings.filterwarnings("ignore")

# Load built-in Airline dataset
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv',
                 parse_dates=['Month'],
                 index_col='Month')

# Rename column for simplicity
df.columns = ['Passengers']

# Plot original data
df.plot(figsize=(10,5), title='Monthly Airline Passengers')
plt.show()

# Train SARIMA model
# (p,d,q) = (1,1,1)
# (P,D,Q,s) = (1,1,1,12) -> seasonal monthly data
model = SARIMAX(df['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

results = model.fit()

# Forecast next 12 months
forecast = results.forecast(steps=12)

# Create future dates
future_dates = pd.date_range(start=df.index[-1], periods=13, freq='M')[1:]

forecast_df = pd.DataFrame({'Forecast': forecast.values}, index=future_dates)

# Plot results
plt.figure(figsize=(10,5))
plt.plot(df, label='Original Data')
plt.plot(forecast_df, label='Forecast', color='red')

plt.title('SARIMA Forecast (Next 12 Months)')
plt.xlabel('Date')
plt.ylabel('Passengers')
plt.legend()
plt.show()

# Print forecast values
print(forecast_df)
```

```python
#Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
#(Include your Python code and output in the code box below.)

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler

# Create synthetic dataset (2D numerical data)
np.random.seed(42)

# Normal data
X1 = np.random.normal(50, 5, 200)
X2 = np.random.normal(50, 5, 200)

# Combine into dataframe
df = pd.DataFrame({'Feature1': X1, 'Feature2': X2})

# Add anomalies manually
df.loc[::20, 'Feature1'] = df['Feature1'] * 2
df.loc[::20, 'Feature2'] = df['Feature2'] * 0.3

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
df['anomaly'] = lof.fit_predict(X_scaled)

# Convert labels (-1 = anomaly, 1 = normal)
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})

# Print anomaly count
print("Total anomalies detected:", df['anomaly'].sum())

# Visualization
plt.figure(figsize=(8,6))

# Normal points
plt.scatter(df[df['anomaly'] == 0]['Feature1'],
            df[df['anomaly'] == 0]['Feature2'],
            label='Normal', alpha=0.6)

# Anomalies
plt.scatter(df[df['anomaly'] == 1]['Feature1'],
            df[df['anomaly'] == 1]['Feature2'],
            color='red', label='Anomaly')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('LOF - Anomaly Detection')
plt.legend()
plt.show()
```

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error

# -----------------------------
# Create synthetic 15-min interval data
# -----------------------------
np.random.seed(42)
date_range = pd.date_range(start='2024-01-01', periods=500, freq='15min')

df = pd.DataFrame({
    'timestamp': date_range,
    'energy_usage': np.random.normal(100, 10, 500),
    'temperature': np.random.normal(25, 5, 500)
})

df.set_index('timestamp', inplace=True)

# Add anomalies (spikes)
df.loc[df.index[::50], 'energy_usage'] *= 1.8

# -----------------------------
# Feature Engineering
# -----------------------------
df['hour'] = df.index.hour
df['dayofweek'] = df.index.dayofweek

# -----------------------------
# Anomaly Detection (Isolation Forest)
# -----------------------------
model_if = IsolationForest(contamination=0.05, random_state=42)

features = ['energy_usage', 'temperature', 'hour', 'dayofweek']
df['anomaly'] = model_if.fit_predict(df[features])

# Convert labels (1 = normal, -1 = anomaly)
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})

print("Total anomalies detected:", df['anomaly'].sum())

# -----------------------------
# Handle anomalies (IMPORTANT FIX)
# Replace anomalies instead of deleting
# -----------------------------
df_clean = df.copy()
df_clean.loc[df_clean['anomaly'] == 1, 'energy_usage'] = np.nan

# Fill missing values using interpolation
df_clean['energy_usage'] = df_clean['energy_usage'].interpolate()

# -----------------------------
# Train-Test Split
# -----------------------------
train = df_clean['energy_usage'].iloc[:400]
test = df_clean['energy_usage'].iloc[400:]

train_exog = df_clean[['temperature']].iloc[:400]
test_exog = df_clean[['temperature']].iloc[400:]

# -----------------------------
# Forecasting (SARIMAX)
# -----------------------------
model_sarimax = SARIMAX(
    train,
    order=(1,1,1),
    seasonal_order=(1,1,1,96),  # daily seasonality for 15-min data
    exog=train_exog,
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model_sarimax.fit(disp=False)

# Model summary (optional but useful)
print(results.summary())

forecast = results.forecast(steps=len(test), exog=test_exog)

# -----------------------------
# Evaluation
# -----------------------------
mae = mean_absolute_error(test, forecast)
print("MAE:", mae)

# -----------------------------
# Visualization
# -----------------------------
plt.figure(figsize=(12,6))

# Actual vs Forecast
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, forecast, label='Forecast')

# Highlight anomalies (original data)
anomalies = df[df['anomaly'] == 1]
plt.scatter(anomalies.index, anomalies['energy_usage'],
            label='Anomaly')

plt.title('Energy Forecasting + Anomaly Detection (Final)')
plt.xlabel('Time')
plt.ylabel('Energy Usage')
plt.legend()
plt.show()
```
